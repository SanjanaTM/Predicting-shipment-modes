```python
#importing necessary packages
import pandas as pd #for creating and handling datasframes
import numpy as np
from mlxtend.feature_selection import SequentialFeatureSelector
data_eda = pd.read_csv(r"C:\Users\LENOVO\Downloads\archive (4)\Train.csv") #for eda purposes
data_pda = pd.read_csv(r"C:\Users\LENOVO\Downloads\archive (4)\Train.csv") #for pda purposes

print('[info] data loaded successfully')
```

```python
import numpy as np
State_Name_unique = data_pda['Warehouse_block'].unique().tolist() #getting a list of all unique values in State_Name column
State_Name_index = np.arange(1,len(State_Name_unique)+1,1) #creating an array fo numbers as indexing with length of above list
State_Name_Next_to_num = {i:j for i,j in zip(State_Name_unique,State_Name_index)} #creating a dictionary for mapping text to num

#mapping text labels in State_Name columns to number
data_pda['Warehouse_block'] =data_pda['Warehouse_block'].map(State_Name_Next_to_num)

#displayig new updated dataframe
data_pda.head(3)
```

```python
import numpy as np
State_Name_unique = data_pda['Mode_of_Shipment'].unique().tolist() #getting a list of all unique values in State_Name column
State_Name_index = np.arange(0,len(State_Name_unique),1) #creating an array fo numbers as indexing with length of above list
State_Name_Next_to_num = {i:j for i,j in zip(State_Name_unique,State_Name_index)} #creating a dictionary for mapping text to num

#mapping text labels in State_Name columns to number
data_pda['Mode_of_Shipment'] =data_pda['Mode_of_Shipment'].map(State_Name_Next_to_num)
print(data_pda['Mode_of_Shipment'].unique())
#displayig new updated dataframe
data_pda.head(3)
```

```python
import numpy as np
State_Name_unique = data_pda['Product_importance'].unique().tolist() #getting a list of all unique values in State_Name column
State_Name_index = np.arange(1,len(State_Name_unique)+1,1) #creating an array fo numbers as indexing with length of above list
State_Name_Next_to_num = {i:j for i,j in zip(State_Name_unique,State_Name_index)} #creating a dictionary for mapping text to num

#mapping text labels in State_Name columns to number
data_pda['Product_importance'] =data_pda['Product_importance'].map(State_Name_Next_to_num)

#displayig new updated dataframe
data_pda.head(3)
```

```python
import numpy as np
State_Name_unique = data_pda['Gender'].unique().tolist() #getting a list of all unique values in State_Name column
State_Name_index = np.arange(1,len(State_Name_unique)+1,1) #creating an array fo numbers as indexing with length of above list
State_Name_Next_to_num = {i:j for i,j in zip(State_Name_unique,State_Name_index)} #creating a dictionary for mapping text to num

#mapping text labels in State_Name columns to number
data_pda['Gender'] =data_pda['Gender'].map(State_Name_Next_to_num)

#displayig new updated dataframe
data_pda.head(3)
```

```python
necessary_columns = ['Warehouse_block','Mode_of_Shipment','Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Product_importance','Gender','Discount_offered','Weight_in_gms','Reached.on.Time_Y.N'] #creating a list of wanted columns
data_pda = data_pda[necessary_columns] #updating the dataframe to contain only necessary columns
data_pda.head() #displaying new dataframe
```

```python
from sklearn.tree import DecisionTreeClassifier #importing the algo
```

```python
#segregating data into features and target
x_columns = ['Warehouse_block','Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Product_importance','Gender','Discount_offered','Weight_in_gms','Reached.on.Time_Y.N'] #definig input columns
x = data_pda[x_columns].values #choosing input columns

y = data_pda['Mode_of_Shipment'].values #choosing output column

#splitting data into training and testing partitions
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2800,test_size=0.1)

print('[info] data segregated and splitting complete')
```

```python
x_train.shape
```

```python
#data = data.dropna()
from sklearn.tree import DecisionTreeClassifier #importing the algo
dt_model = DecisionTreeClassifier() #initializing the model
dt_model.fit(x_train,y_train) #training the model on training data
print('[info] model training complete...')
```

```python
#using the trained model to predict on test input data(x_test)
dt_model_predicted_values = dt_model.predict(x_test)

#defining actual values
dt_model_actual_values = y_test

#comparing actual values with predicted vlues to get accuracy, precision,etc.,
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

accuracy = accuracy_score(dt_model_actual_values,dt_model_predicted_values)
print('accuracy of the model is : ',accuracy)

#precision = precision_score(dt_model_actual_values,dt_model_predicted_values)
#print('precision of the model is : ',precision)

#recall = recall_score(dt_model_actual_values,dt_model_predicted_values)
#print('recall of the model is : ',recall)

#f1 = f1_score(dt_model_actual_values,dt_model_predicted_values)
#print('f1 score of the model is : ',f1)
```

```python
from sklearn.metrics import classification_report
print('classification report is : \n',classification_report(dt_model_actual_values,dt_model_predicted_values))
```

```python
#creating confusion matrix from trained model
from sklearn.metrics import confusion_matrix
dt_model = confusion_matrix(dt_model_actual_values,dt_model_predicted_values)

print(dt_model) #displaying confusion matrix

#displaying confusion matrix as heatmap
import seaborn as sns
sns.heatmap(dt_model)
```
